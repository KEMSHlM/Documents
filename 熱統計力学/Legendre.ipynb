```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x0 = 1
y0 = 1
u0 = 2 * x0
v0 = 2 * y0

color = 'black'
```

```python
# x, y, u, v
x = y = u = v = np.linspace(0, 5, 100)
x, y = np.meshgrid(x, y)
u, v = np.meshgrid(u, v)

# 
z1 = x**2 + y**2
z2 = -u**2/4 + y**2
z3 = x**2 - v**2/4
z4 = - (u**2 + v**2) / 4

# 3D
fig = plt.figure(figsize=(20, 5))  # figsize

#  (x, y) = x^2 + y^2
ax1 = fig.add_subplot(141, projection='3d')  # 
ax1.plot_surface(x, y, z1, cmap='viridis')
ax1.set_title('$\phi (x, y) = x^2 + y^2$')

#  (u, y) = - u^2/4 + y^2
ax2 = fig.add_subplot(142, projection='3d')  # 
ax2.plot_surface(u, y, z2, cmap='viridis')
ax2.set_title('$\psi (u, y) = - u^2/4 + y^2$')

#  (x, v) = x^2 - v^2/4
ax3 = fig.add_subplot(143, projection='3d')  # 
ax3.plot_surface(x, v, z3, cmap='viridis')
ax3.set_title('$\lambda (x, v) = x^2 - v^2/4$')

#  (u, v) = - (u^2 + v^2) / 4
ax4 = fig.add_subplot(144, projection='3d')  # 
ax4.plot_surface(u, v, z4, cmap='viridis')
ax4.set_title('$\mu (u, v) = - (u^2 + v^2) / 4$')

# 
# ax1.scatter(x0, y0, x0**2 + y0**2, color=color)
# ax2.scatter(u0, y0, - u0**2/4 + y0**2, color=color)
# ax3.scatter(x0, v0, x0**2 - v0**2/4, color=color)
# ax4.scatter(u0, v0, - (u0**2 + v0**2) / 4, color=color)
```
